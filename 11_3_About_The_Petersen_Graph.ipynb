```
import pydot
import graphviz

dot_string = """graph petersen_graph {

    node [
      fontname = "Arial"
      shape = "circle"
      width = "0.400000"
      height = "0.400000"
      color = "black"
    ]
    edge [
      color = "black"
    ]

    f -- h -- j -- g -- i -- f

    a -- b -- c -- d -- e -- a

    a -- f [weight = f]
    b -- g [weight = f]
    c -- h [weight = f]
    d -- i [weight = f]
    e -- j [weight = f]

    

}"""

graph = pydot.graph_from_dot_data(dot_string)[0]
graphviz.Source(graph)
```

```
import networkx as nx

nx.draw(nx.petersen_graph())
```
